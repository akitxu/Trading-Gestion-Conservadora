```python

# ============================================================
# ADVERTENCIA PARA MOSTRAR EN PANTALLA, EN HTML Y EN EL PDF
# ============================================================

advertencia_texto = (
    "ADVERTENCIA IMPORTANTE PARA INVERSORES NO PROFESIONALES\n\n"
    "Los escenarios hist√≥ricos y las predicciones generadas mediante modelos "
    "estad√≠sticos o t√©cnicas de inteligencia artificial no constituyen garant√≠as "
    "de resultados futuros. Estos modelos se basan exclusivamente en datos pasados "
    "y no pueden anticipar cambios estructurales del mercado, eventos inesperados "
    "o crisis financieras. Las predicciones deben interpretarse como herramientas "
    "orientativas, no como recomendaciones de inversi√≥n.\n\n"
    "Toda decisi√≥n debe considerar el perfil de riesgo, la diversificaci√≥n, el "
    "horizonte temporal y la capacidad de asumir p√©rdidas. La inversi√≥n en mercados "
    "financieros implica riesgos, incluyendo la p√©rdida total del capital invertido."
)

# Mostrar en pantalla
#print(advertencia_texto)

# Mostrar en HTML dentro del notebook
from IPython.display import HTML, display

display(HTML(f"""
<div style="
    border: 2px solid #b30000;
    padding: 20px;
    border-radius: 10px;
    background-color: #ffe6e6;
    font-family: Arial, sans-serif;
    font-size: 14px;
    line-height: 1.5;
">
<h2 style="color:#b30000; text-align:center;">ADVERTENCIA IMPORTANTE PARA INVERSORES NO PROFESIONALES</h2>
<p>
Los escenarios hist√≥ricos y las predicciones generadas mediante modelos estad√≠sticos o t√©cnicas de inteligencia artificial no constituyen garant√≠as de resultados futuros. Estos modelos se basan exclusivamente en datos pasados y no pueden anticipar cambios estructurales del mercado, eventos inesperados o crisis financieras. Las predicciones deben interpretarse como herramientas orientativas, no como recomendaciones de inversi√≥n.
</p>
<p>
Toda decisi√≥n debe considerar el perfil de riesgo, la diversificaci√≥n, el horizonte temporal y la capacidad de asumir p√©rdidas. La inversi√≥n en mercados financieros implica riesgos, incluyendo la p√©rdida total del capital invertido.
</p>
</div>
"""))


# ============================================================
# Importaci√≥n de datos desde Yahoo Finance
# ============================================================

def descargar_serie_yahoo(ticker, start="2010-01-01", end=None):
    import datetime as dt
    import pandas as pd
    import yfinance as yf

    if end is None:
        end = dt.date.today().strftime("%Y-%m-%d")

    data = yf.download(ticker, start=start, end=end, progress=False, auto_adjust=False)

    if data is None or data.empty:
        return None

    if isinstance(data.columns, pd.MultiIndex):
        data.columns = [col[0] for col in data.columns]

    if "Close" not in data.columns:
        print(f"‚ö†Ô∏è No se encontr√≥ columna 'Close' para {ticker}")
        print("Columnas disponibles:", list(data.columns))
        return None

    df = data[["Close"]].copy()
    df = df.rename(columns={"Close": "precio"})
    df.index.name = "fecha"
    df = df.reset_index()

    df["precio"] = pd.to_numeric(df["precio"], errors="coerce")
    df = df.dropna()

    return df


df = descargar_serie_yahoo('^IBEX', start="2010-01-01")


# ============================================================
# Predicci√≥n Prophet + LSTM + Monte Carlo (GARCH)
# ============================================================

import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

from prophet import Prophet

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras import Input
from sklearn.preprocessing import MinMaxScaler

from arch import arch_model

from pathlib import Path

# Crear directorio de informes si no existe
ruta_informes = Path("Informes_predicciones")
ruta_informes.mkdir(parents=True, exist_ok=True)

# ============================================================
# 1. Predicci√≥n con Prophet
# ============================================================

def predecir_prophet(df, dias_futuro=252):
    df_p = df.rename(columns={"fecha": "ds", "precio": "y"})
    modelo = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False
    )
    modelo.fit(df_p)
    futuro = modelo.make_future_dataframe(periods=dias_futuro)
    forecast = modelo.predict(futuro)
    return forecast


# ============================================================
# 2. Predicci√≥n con LSTM (corregida)
# ============================================================

def predecir_lstm(df, pasos=60, dias_futuro=60):
    precios = df["precio"].values.reshape(-1, 1)

    scaler = MinMaxScaler()
    precios_scaled = scaler.fit_transform(precios)

    X, y = [], []
    for i in range(pasos, len(precios_scaled)):
        X.append(precios_scaled[i-pasos:i, 0])
        y.append(precios_scaled[i, 0])

    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))

    model = Sequential([
        Input(shape=(X.shape[1], 1)),
        LSTM(50, return_sequences=True),
        LSTM(50),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    model.fit(X, y, epochs=20, batch_size=32, verbose=0)

    pred_input = precios_scaled[-pasos:].reshape(1, pasos, 1)
    predicciones = []

    for _ in range(dias_futuro):
        pred = model.predict(pred_input, verbose=0)[0][0]
        predicciones.append(pred)

        pred_reshaped = np.array(pred).reshape(1, 1, 1)
        pred_input = np.concatenate([pred_input[:, 1:, :], pred_reshaped], axis=1)

    predicciones = scaler.inverse_transform(np.array(predicciones).reshape(-1, 1))
    return predicciones.flatten()


# ============================================================
# 3. Monte Carlo GARCH (versi√≥n robusta)
# ============================================================

def simular_garch(df, dias_futuro=252, n_sim=2000):
    precios = df["precio"]
    retornos = precios.pct_change().dropna()

    modelo = arch_model(retornos * 100, vol="Garch", p=1, q=1, dist="normal")
    res = modelo.fit(disp="off")

    omega = res.params["omega"]
    alpha = res.params["alpha[1]"]
    beta  = res.params["beta[1]"]

    ultimo_ret = retornos.iloc[-1] * 100
    ultimo_sigma2 = res.conditional_volatility.iloc[-1]**2

    sims = np.zeros((n_sim, dias_futuro))
    precios_simulados = np.zeros((n_sim, dias_futuro))

    precio_inicial = precios.iloc[-1]

    for i in range(n_sim):
        ret = ultimo_ret
        sigma2 = ultimo_sigma2
        precio = precio_inicial

        for t in range(dias_futuro):
            sigma2 = omega + alpha * (ret**2) + beta * sigma2
            sigma = np.sqrt(sigma2)

            eps = np.random.normal(0, sigma)
            ret = eps
            sims[i, t] = ret

            precio = precio * (1 + ret/100)
            precios_simulados[i, t] = precio

    return precios_simulados


# ============================================================
# 4. EJECUCI√ìN DE LOS MODELOS
# ============================================================

dias_futuro = 252

print("Calculando Prophet...")
forecast_prophet = predecir_prophet(df, dias_futuro)

print("Calculando LSTM...")
pred_lstm = predecir_lstm(df, pasos=60, dias_futuro=60)

print("Simulando Monte Carlo GARCH...")
sims = simular_garch(df, dias_futuro, n_sim=2000)

p5 = np.percentile(sims, 5, axis=0)
p50 = np.percentile(sims, 50, axis=0)
p95 = np.percentile(sims, 95, axis=0)


# ============================================================
# 5. EXPORTACI√ìN A PDF + MOSTRAR GR√ÅFICOS
# ============================================================

with PdfPages(ruta_informes /"predicciones_IA.pdf") as pdf:

    # P√°gina 1: Advertencia
    fig_adv = plt.figure(figsize=(11.7, 8.3))
    plt.axis("off")
    plt.text(0.5, 0.5, advertencia_texto, ha="center", va="center", fontsize=12, wrap=True)
    pdf.savefig(fig_adv)
    plt.close(fig_adv)

    # Prophet
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(df["fecha"], df["precio"], label="Hist√≥rico")
    ax.plot(forecast_prophet["ds"], forecast_prophet["yhat"], label="Predicci√≥n Prophet")
    ax.fill_between(forecast_prophet["ds"], forecast_prophet["yhat_lower"],
                    forecast_prophet["yhat_upper"], alpha=0.3)
    ax.set_title("Predicci√≥n Prophet")
    ax.grid(True)
    ax.legend()
    pdf.savefig(fig)
    plt.show()
    plt.close(fig)

    # LSTM
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(df["fecha"], df["precio"], label="Hist√≥rico")
    fechas_futuras = pd.date_range(df["fecha"].iloc[-1], periods=len(pred_lstm)+1, freq="D")[1:]
    ax.plot(fechas_futuras, pred_lstm, label="Predicci√≥n LSTM")
    ax.set_title("Predicci√≥n LSTM")
    ax.grid(True)
    ax.legend()
    pdf.savefig(fig)
    plt.show()
    plt.close(fig)

    # Monte Carlo
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(df["fecha"], df["precio"], label="Hist√≥rico")
    fechas_mc = pd.date_range(df["fecha"].iloc[-1], periods=dias_futuro+1, freq="D")[1:]
    ax.plot(fechas_mc, p50, label="Mediana MC")
    ax.fill_between(fechas_mc, p5, p95, alpha=0.3)
    ax.set_title("Simulaci√≥n Monte Carlo (GARCH)")
    ax.grid(True)
    ax.legend()
    pdf.savefig(fig)
    plt.show()
    plt.close(fig)

    # Comparativa
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(df["fecha"], df["precio"], label="Hist√≥rico")
    ax.plot(forecast_prophet["ds"], forecast_prophet["yhat"], label="Prophet")
    ax.plot(fechas_futuras, pred_lstm, label="LSTM")
    ax.plot(fechas_mc, p50, label="Monte Carlo (mediana)")
    ax.set_title("Comparativa de m√©todos de predicci√≥n")
    ax.grid(True)
    ax.legend()
    pdf.savefig(fig)
    plt.show()
    plt.close(fig)

print("PDF generado: predicciones_IA.pdf")
```
